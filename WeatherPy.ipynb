```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import date

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy # installed locally 12/15/19

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
zip_lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
zip_lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in zip_lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
        
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# # Identify corresponding country code
# for city in cities:
#     country_code = city.country_code
#     countries.append(country_code)    

    
# Print the city count to confirm sufficient count
len(cities)
# countries
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# URL Parameters
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# URL Builder
query_url = base_url + "appid=" + api_key + "&units=" + units + "&q="
```

```python
# Data Storage bins
name = [] # city name
country = [] # abbreviated
lat = []
lng = []
temperature = [] # Fahrenheit
humidity = [] # Percent
cloudiness = [] # Percent
wind_speed = [] # MPH

# Epic City Weather Loop
for num, city in enumerate(cities):
    try:
        weather_report = requests.get(query_url + city).json()
        name.append(weather_report['name'])
        country.append(weather_report['sys']['country'])
        lat.append(weather_report['coord']['lat'])
        lng.append(weather_report['coord']['lon'])
        temperature.append(weather_report['main']['temp'])
        humidity.append(weather_report['main']['humidity'])
        cloudiness.append(weather_report['clouds']['all'])
        wind_speed.append(weather_report['wind']['speed'])
        # pause loop for API limit
        time.sleep(1)
        # print each json lookup
        print(f"[|Item {num}|] Pulling city weather data: {city}")
        print("----------")
    except KeyError:
        print(f"[|Item {num}|] >> Whoops. Nothing here! <<next>>")
        print("----------")

print("==============")
print("*GREAT NEWS!* Data retrieval complete.")
#
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
import_dict = {
    "City":name,
    "Country":country,
    "Latitude":lat,
    "Longitude":lng,
    "Temperature (F)":temperature,
    "Humidity (%)":humidity,
    "Cloudiness (%)":cloudiness,
    "Wind Speed (mph)":wind_speed}

equator_proof = pd.DataFrame(import_dict)

# Export to CSV
equator_proof.to_csv("weather_report.csv")
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.

#### Latitude vs. Temperature Plot

```python
lat_vs_temp = plt.scatter(equator_proof["Latitude"], equator_proof["Temperature (F)"], marker="o", alpha=.5)
plt.grid(axis="both")
plt.title(f"Latitude vs. Temperature Plot ({date.today()})")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
lat_vs_temp

plt.savefig("images/Latitude_vs_Temperature.png")
```

#### Latitude vs. Humidity Plot

```python
lat_vs_hum = plt.scatter(equator_proof["Latitude"], equator_proof["Humidity (%)"], marker="o", alpha=.5)
plt.grid(axis="both")
plt.title(f"Latitude vs. Humidity Plot ({date.today()})")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
lat_vs_hum

plt.savefig("images/Latitude_vs_Humidity.png")
```

#### Latitude vs. Cloudiness Plot

```python
lat_vs_cloud = plt.scatter(equator_proof["Latitude"], equator_proof["Cloudiness (%)"], marker="o", alpha=.5)
plt.grid(axis="both")
plt.title(f"Latitude vs. Cloudiness Plot ({date.today()})")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
lat_vs_cloud

plt.savefig("images/Latitude_vs_Clouds.png")
```
